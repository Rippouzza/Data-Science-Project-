```python
import cv2
import torch
import easyocr
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import re

# Initialisation d'EasyOCR
reader = easyocr.Reader(['en'], gpu=True)

# Chargement du modÃ¨le YOLOv5
model_path = '/content/drive/MyDrive/DSproject/yolov5/runs/train/exp/weights/best.pt'
try:
    model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)
except Exception as e:
    print(f"Erreur lors du chargement du modÃ¨le YOLOv5: {e}")
    exit()

# Charger la vidÃ©o
video_path = '/content/drive/MyDrive/DSproject/converted2_video.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Erreur : Impossible d'ouvrir la vidÃ©o.")
    exit()

# DÃ©tails de la vidÃ©o
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = '/content/drive/MyDrive/DSproject/output2_video.mp4'
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# PrÃ©traitement pour EasyOCR
def preprocess_for_ocr(image):
    # Convertir en niveaux de gris
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # AmÃ©liorer le contraste
    enhanced = cv2.equalizeHist(gray)

    # Appliquer un seuil pour rendre l'image binaire
    _, binary = cv2.threshold(enhanced, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Redimensionner pour augmenter la taille de l'image
    resized = cv2.resize(binary, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

    # Appliquer un flou pour lisser l'image
    blurred = cv2.GaussianBlur(resized, (5, 5), 0)

    # Retourner l'image prÃ©traitÃ©e
    return blurred

# Fonction pour vÃ©rifier si le texte contient uniquement des chiffres
def extract_numbers(text):
    numbers = re.findall(r'\d+', text)  # Trouver tous les nombres dans le texte
    return ' '.join(numbers) if numbers else ""  # Retourner les nombres trouvÃ©s

# Traiter chaque frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # DÃ©tection d'objets YOLOv5
    results = model(frame)
    detections = results.xyxy[0].cpu().numpy()

    for det in detections:
        x1, y1, x2, y2, conf, cls = map(int, det[:6])

        # Ajouter une marge Ã  la boÃ®te
        margin = 5
        x1 = max(0, x1 - margin)
        y1 = max(0, y1 - margin)
        x2 = min(frame.shape[1], x2 + margin)
        y2 = min(frame.shape[0], y2 + margin)

        if y2 <= y1 or x2 <= x1:
            continue

        # PrÃ©parer la ROI pour l'OCR
        roi = frame[y1:y2, x1:x2]
        processed_roi = preprocess_for_ocr(roi)

        # DÃ©tection de texte avec EasyOCR
        ocr_result = reader.readtext(processed_roi, detail=0, paragraph=True)
        detected_text = ' '.join(ocr_result) if ocr_result else ""

        # Filtrer pour ne conserver que les chiffres
        detected_numbers = extract_numbers(detected_text)

        # Ignorer si aucun chiffre n'est dÃ©tectÃ©
        if not detected_numbers:
            continue

        # Ajouter "ØªÙˆÙ†Ø³" aprÃ¨s les chiffres dÃ©tectÃ©s
        custom_message = f"{detected_numbers} ØªÙˆÙ†Ø³"

        # Convertir l'image OpenCV en image PIL pour ajouter le texte
        pil_img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        draw = ImageDraw.Draw(pil_img)

        # Utiliser une police qui supporte l'arabe
        try:
            font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 30)
        except IOError:
            font = ImageFont.truetype("/content/drive/MyDrive/DSproject/fonts/Rubik-VariableFont_wght.ttf", 30)  # Ajustez le chemin vers la police arabe

        # Ajouter le texte arabe et les chiffres dÃ©tectÃ©s
        draw.text((x1, y2 + 20), custom_message, font=font, fill=(0, 255, 0))

        # Convertir l'image PIL en OpenCV
        frame = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

        # Dessiner la boÃ®te de dÃ©tection
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # Ã‰crire la frame dans la vidÃ©o de sortie
    out.write(frame)

# LibÃ©rer les ressources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"VidÃ©o traitÃ©e enregistrÃ©e dans : {output_path}")
```
