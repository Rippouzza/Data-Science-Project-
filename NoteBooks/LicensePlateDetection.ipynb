```python
import yaml

# Define the correct paths
train_path = '/content/drive/MyDrive/dataset projet/train/images'
val_path = '/content/drive/MyDrive/dataset projet/valid/images'
test_path = '/content/drive/MyDrive/dataset projet/test/images'

# Path to the YAML file
yaml_path = '/content/drive/MyDrive/dataset projet/data.yaml'

# Load and update the YAML file
with open(yaml_path, 'r') as file:
    data = yaml.safe_load(file)

data['train'] = train_path
data['val'] = val_path
data['test'] = test_path

# Save the updated YAML file
with open(yaml_path, 'w') as file:
    yaml.dump(data, file)

print("Updated data.yaml successfully!")
```

```python
with open(yaml_path, 'r') as file:
    updated_data = yaml.safe_load(file)
print(updated_data)
```
